```python
import numpy as np
from numpy import linalg as la
from HubbardClass import hubbard
import csv
import scipy.linalg
from itertools import permutations 

#First I define the chain length, possible fillings, temperature, chemical potential and on site energies
L = 4
N = np.arange(0,L+1,1)
beta = 1
temp = 1/beta
mu = 0.1

runs=500
#Now create array to store equilibrium densities and value of grand potential
ni = np.zeros((runs,L+1))

#unfortunately the nhats will change depending on N, so I will just define these within the loops
#outer loop is for each set of final densities
for i in range(runs):
  energies = np.random.rand(L)*2
  Z=0
  #now loop over possible particle numbers
  for j in range(len(N)):
    datagen = hubbard(sites = L, filling = N[j], pbcs = True, t=1, eps = energies, V = 3, gridtype='1D')
    Ham = datagen.hamiltonian()
    HD = la.eigh(Ham)  
    evec = HD[1].T 
    eval = HD[0]
    d = len(HD[0])
    #Now generate the number operators for this given filling
    nhat = np.zeros((L,d,d))
    for p in range(L):
      for q in range(d):
        if(datagen.isKthBitSet(int(datagen.states[q]),p+1)):
          nhat[p][q][q]+=1
    
    ##now have to calculate the density that will be used in the problem, will divide by partition function in outer loop
    for l,n in enumerate(nhat):
      for k in range(d):
        ni[i][l]+=evec[k].dot(n.dot(evec[k]))*np.exp(-beta*eval[k])

    #the number operator for the system is N*identity operator
    numberop = np.zeros((d,d))
    for r in range(d):
      numberop[r][r]=N[j]

    #grand partition function is sum of Z*fugacity
    X=Z
    Z += np.trace(scipy.linalg.expm(-beta*(Ham-mu*numberop)))

  #now I can divide by Z and add the grand potential at the end
  grand = -temp*np.log(Z)
  ni[i] = ni[i]/Z
  ni[i][L] += grand-np.dot(energies,ni[i,0:L])
  #print(i)

with open(f"/content/drive/MyDrive/Final Year Project/Data/grandcanonical/84trainingdata.csv", 'w', newline='') as myFile:
    writer = csv.writer(myFile)
    writer.writerows(ni)
```
